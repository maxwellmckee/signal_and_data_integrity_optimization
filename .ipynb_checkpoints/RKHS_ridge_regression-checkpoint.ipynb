#### Overview: RKHS Ridge Regression
This notebook involves denoising an image contained within `peaks.mat` through the application of Reproducing Kernel Hilbert Space (RKHS) techniques, specifically utilizing Ridge Regression. The steps include:

* Constructing a 2D Gaussian kernel basis (Gram matrix) by calculating the Kronecker product of two 1D Gaussian kernels with a specified bandwidth.
* Implementing the RKHS Ridge Regression procedure using the constructed Gram matrices to estimate the value of each pixel based on the surrounding pixel values.
* Calculating the smooth version of the given image and estimating the noise's standard deviation, followed by visualizing the denoised image.

#### Image Denoising with RKHS and Ridge Regression
(a) We initiate the denoising process by generating a 2D Gaussian kernel basis. This is achieved by computing the Kronecker product between two 1D Gaussian kernels, each with a bandwidth setting of 1.

(b) With the 2D Gaussian kernel basis ready, we proceed to refine the pixel values across the noisy image. Then the RKHS Ridge Regression methodology is applied, leveraging the spatial information encoded in the Gram matrices to enhance each pixel's value.

(c) Lastly, we finalize the task by both deducing and illustrating the noise-reduced image. Additionally, we undertake an analysis to determine the standard deviation of the noise that affected the original image, complementing the denoising process with a quantitative noise level assessment.

### Start denoising process by creating Gaussian kernel basis
First we define a function to compute the Gaussian kernel between two input arrays. `cdist` is used to compute the squared Euclidean distance between every pair of points created by the two input arrays. Then, we create a 1D Gaussian kernel matrix by applying the function to the `x` array with itself. Lastly, we find the 2D Gaussian kernel basis (aka Gram matrix) by taking the Kronecker (`kron`) product of the 1D Gaussian kernel with itself. The result is a 2D Gaussian kernel basis of shape (10000,10000) which represents the image coordinates of a $100 \times 100$ image. The entries of this basis represent the similarity between different points in the image.

```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from scipy.linalg import kron

data = sio.loadmat('peaks.mat')
Y = data['Y']
print(f'image shape: {Y.shape}')
plt.imshow(Y, cmap='gray')
plt.axis('off')
plt.title('Noisy Image')
plt.show()
```

```python
# Create a 1D array representing the image coordinates along one dimension
x = np.linspace(0, 1, 100).reshape(-1, 1)

# Define the Gaussian kernel function with bandwidth 1
def gaussian_kernel(x1, x2, bandwidth=1):
    return np.exp(-0.5 * cdist(x1, x2, 'sqeuclidean') / bandwidth**2)

# Create the 1D Gaussian kernel
Gaussian_1D = gaussian_kernel(x, x)

# Create the 2D Gaussian kernel basis (Gram matrix) using the Kronecker product
Gaussian_2D = kron(Gaussian_1D, Gaussian_1D)

print(f'Gaussian kernel shape: {Gaussian_2D.shape}')
```

### Refine pixel values
RKHS involves solving a linear system using the Gram matrix. In order to estimate the pixel values in the image, we can use the ridge regression formula: $\alpha = (K + \lambda * I)^{-1} * y$ where $K$ is the Gram matrix. When solving for an optimal ridge parameter, I get a warning about an Ill-conditioned matrix leading to likely inoptimal $\lambda$ parameters. You can potentiall help ensure the matrix is invertible by adding a small constant to the diagonal elements of the matrix before inverting it (I guess this is called "jittering"). Then, the estimated pixel values `Y_hat` are computed by multiplying the Gram matrix by the alpha coefficients, leading to a denoised version of the image.

```python
from sklearn.linear_model import RidgeCV

# Define the range of lambda values you want to test
lambda_values = np.logspace(-10, 10, 10)  # for example

# Create a RidgeCV object, specifying the lambda values and the cross-validation strategy
ridge = RidgeCV(alphas=lambda_values, cv=5)  # 5-fold cross-validation

# Fit the model to the data (you'll need to reshape Y into a 1D array)
ridge.fit(Gaussian_2D, Y.reshape(-1))

# The optimal lambda value is stored in the `alpha_` attribute
optimal_lambda = ridge.alpha_

print("Optimal lambda:", optimal_lambda)
```

```python
from numpy.linalg import inv

# Reshape Y to a 1D array
Y_1D = Y.reshape(-1)

# Jitter value
jitter = 1e-10  # for example

# Compute alpha
alpha = np.linalg.solve(Gaussian_2D + optimal_lambda * np.eye(Gaussian_2D.shape[0]) + jitter * np.eye(Gaussian_2D.shape[0]), Y_1D)

# Compute the estimated pixel values
Y_hat = np.dot(Gaussian_2D, alpha)

# Reshape Y_hat back to the original shape
Y_hat = Y_hat.reshape(Y.shape)
print(f'Y_hat shape: {Y_hat.shape}')
```

### Deduce and illustrate noise-reduced image
The image noise is the difference between the original noisy image and `Y` and the estimated smooth image `Y_hat`. The noise standard deviation is then estimated as the standard deviation of the noisy image.

```python
# Compute the noise image
noise = Y - Y_hat

# Estimate the noise standard deviation
noise_std = np.std(noise)

print("Estimated noise standard deviation:", noise_std)

# Plot the smooth image
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(Y_hat, cmap='gray')
plt.title("Smooth Image")
plt.colorbar()

# Plot the noise image
plt.subplot(1, 2, 2)
plt.imshow(noise, cmap='gray')
plt.title("Noisy Image")
plt.colorbar()

plt.tight_layout()
plt.show()
```
