#### Overview: Smooth Sparse Decomposition
This centers around the concept of smooth-sparse decomposition focusing on an optimization problem designed to decompose data into its smooth and sparse components. The decomposition is achieved by minimizing a composite objective function that balances the fidelity of reconstruction, smoothness of the solution, and its sparsity. Specifically, the objective function combines the squared $l_2$ norm of the reconstruction error, a smoothness penalty governed by a Laplacian matrix $\Omega$, and the $l_1$ norm of the sparse component weighted by coefficients $\lambda$ and $\gamma$, respectively. The task involves:

* Demonstrating the convexity of the optimization problem.
* Visualizing the smooth, sparse, and error components of data provided in a data.mat file, effectively separating and illustrating these components in the plotted output.

#### Decomposing Data into Smooth and Sparse Elements
Based on the work of Yan et al., this question invites you to explore an optimization strategy aimed at disentangling a dataset into its smooth and sparse constituents. The objective of this task is defined by the equation:

$$ \min_{\theta, \theta_a} \parallel Y - B\theta - B_a \theta_a \parallel_2^2 + \lambda \theta^T \Omega \theta + \gamma \parallel \theta_a \parallel_1 $$

wherein $\Omega$ represents the Laplacian matrix, expressed as $D^TD$ with $D$ being the difference matrix.

We want to show the formulation is a convex problem. Then, plot the smooth part, anomaly part, and error part of the data.

### Formulation is a convex problem
-------------
To argue that the formulation is a convex problem, we must show that the objective function above is convex. So, let's consider each term of the equation individually.

1. $\parallel Y - B\theta - B_a \theta_a \parallel_2^2$: This term has the Euclidean L2 norm squared, which is a convex function. Convexity is conserved under linear transformations.
2. $\lambda \theta^T \Omega \theta$: This is a quadratic term. The convexity of this term is dependent on if $\Omega$ is positive semi-definite. Since $\Omega = D^TD$, $\Omega$ is positive semi-definite because any real vector $x$, $x^T \Omega x = x^TD^TDx = ||Dx||_2^2 \ge 0$.
3. $\gamma \parallel \theta_a \parallel_1$: This is an L1 norm term that is weighted by a non-negative scalar $\gamma$, which is convex. The L1 norm is a sum of absolute values, and absolute value is a convex function.

Therefore, since the sum of convex terms is a convex term, the objective function is convex.

### Plotting smooth, anomaly, and error

```python
import scipy.io as sio
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt

data = sio.loadmat('data.mat')

B = data['B']
Ba = data['Ba'] 
Omega = data['Omega']
Y = data['Y'].reshape(1000,)
Basis = [B]
Basisa = [Ba]

print(f"B     shape: {B.shape}")
print(f"Ba    shape: {Ba.shape}")
print(f"Omega shape: {Omega.shape}")
print(f"Y     shape: {Y.shape}")
```

```python
# define theta and theta_a as variables
theta = cp.Variable(B.shape[1])
theta_a = cp.Variable(Ba.shape[1])

# define constants lambda and gamma
lambda_ = cp.Parameter(nonneg=True)
gamma = cp.Parameter(nonneg=True)

# set the vals of lambda and gamma
lambda_.value = 0.1
gamma.value = 0.1

# define objective function
objective = cp.Minimize(cp.sum_squares(Y - B@theta - Ba@theta_a) + lambda_*cp.quad_form(theta, Omega) + gamma*cp.norm1(theta_a))

# define constraints
constraints = []

# define problem
problem = cp.Problem(objective, constraints)

# solve problem
problem.solve()

# obtain optimal values
theta_optimal = theta.value
theta_a_optimal = theta_a.value

print(f'theta_optimal  : {theta_optimal}')
print(f'\ntheta_a_optimal: {theta_a_optimal}')
```

```python
# compute yhat and a
yhat = B @ theta_optimal
a = Ba @ theta_a_optimal

# plot yhat
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.plot(yhat)
plt.title('yhat')

# plot a
plt.subplot(1, 3, 2)
plt.plot(a)
plt.title('a')

# plot Y - yhat - a
plt.subplot(1, 3, 3)
plt.plot(Y - yhat - a)
plt.title('Y - yhat - a')

plt.tight_layout()
plt.show()
```
